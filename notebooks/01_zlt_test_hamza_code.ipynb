```python
import numpy as np
import pandas as pd

def validate_normalization(file_path):
    df = pd.read_parquet(file_path)
    min_val = df.min().min()
    max_val = df.max().max()
    if min_val < 0 or max_val > 1:
        print(f"Normalization failed: min={min_val}, max={max_val}")
    else:
        print("Normalization successful: all values within [0, 1]")

# Usage
validate_normalization('../ml/processed_events/normalized_features.parquet')
```
